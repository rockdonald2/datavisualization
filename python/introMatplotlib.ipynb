```python
# tehát létrehozunk egy rácselrendezést, 1 oszloppal és 3 sorral, tehát 3 egymásra helyezett ábránk lesz,
# megosztjuk a 3 ábra között az x és y tengelyeket, tehát normalizáltak lesznek,
# majd beállítjuk a vászon nagyságát, 8x8 inchre
fig, axes = plt.subplots(nrows=3, ncols=1, sharex=True, sharey=True, figsize=(8, 8))

# létrehozunk egy dict-et, a zip segítségével, amely zippeli a random teszt adatainkat a vonaladatokkal,
# tehát, valahogy így fog kinézni: {('y érték', 'foo', 'b'), ...}
# a transzponáltra azért van szükség, mert az y shape-je (200, 3) jelenleg, nekünk megkell fordítanunk ahhoz, hogy
# sor-oszlop legyen, tehát a transzponáltal a shape-je (3, 200)-ra vált
labelled_data = zip(y.transpose(), ('foo', 'bar', 'baz'), ('b', 'g', 'r'))

# címezzük az ábránkat
fig.suptitle('Three random trends', fontsize=16)

# létrehozzuk az ábrákat
for i, ld in enumerate(labelled_data):
    ax = axes[i]
    # használjuk a zip-el létrehozott struktúránkat
    ax.plot(x, ld[0], label=ld[1], color=ld[2])
    ax.set_ylabel('Cum. sum')
    # elhelyezzük a jelmagyarázatot
    ax.legend(loc='upper left', framealpha=0.5, prop={'size': 'small'})
    
# az utolsó ábrát ellátjuk az x tengely címkéjével is, máshova felesleges lenne
axes[-1].set_xlabel('Date')
```

```python
# meghatározzuk az oszlopcímkéket
labels = ['Physics', 'Chemistry', 'Literature', 'Peace']
# az adatokat, ami jelen esetben a magassága lesz az oszlopnak
foo_data = [3, 6, 10, 4]

# oszlopszélesség, számítások során fontos
bar_width = 0.5
# az x tengelyen való elhelyezkedés, azért van a + bar_width, mert így baloldalt kimarad 1 féloszlopnak a helye
xlocations = np.array(range(len(foo_data))) + bar_width

# létrehozzuk az oszlopot, az előbb szerzett elhelyezkedéssel, az adatokkal, és a hosszal
plt.bar(xlocations, foo_data, width=bar_width)

# beállítunk egy ideillő y tengely-címkézést
plt.yticks(range(0, 12))

# elhelyezzük az oszlopok alá a címkét
plt.xticks(xlocations, labels)

# az x tengelyt 0-tól, az utolsó oszlop + 1 féloszlop helye szerinti határra allítjuk
plt.xlim(0, xlocations[-1] + bar_width)

# beállítjuk a címet
plt.title('Prizes won by Fooland')

# meghatározza, hogy hol legyen a tengelyeken a jelölés, lehet az x tengely fent vagy lent,
# az y bal vagy jobb oldalt
plt.gca().get_xaxis().tick_bottom()
plt.gca().get_yaxis().tick_left()

# beállítja a jelenlegi figure méretét
plt.gcf().set_size_inches((8, 4))
```

```python
# meghatározzuk az oszlopcímkéket
labels = ['Physics', 'Chemistry', 'Literature', 'Peace']
# az adatokat, ami jelen esetben a magassága lesz az oszlopnak
foo_data = [3, 6, 10, 4]
bar_data = [8, 3, 6, 1]

fig, ax = plt.subplots(figsize=(8, 4))

# ez az oszlopszélesség kiadja az 1-et, hiszen 0.4 szélesség mellett 0.1 oszlop padding lesz
bar_width = 0.4

# létrehozzuk az elhelyezkedés-rendet
xlocs = np.arange(len(foo_data))
# elhelyezzük az első, világosabb oszlopot
ax.bar(xlocs - bar_width, foo_data, bar_width, color='#fde0bc', label='Fooland')
# elhelyezzük a második, sötétebb oszlopot
ax.bar(xlocs, bar_data, bar_width, color='peru', label='Barland')

# beállítjuk az y tengely tickjeit
ax.set_yticks(range(12))

# manuális határoztam meg a x tengely jelöléseinek elhelyezkedését egy tömb-ben
ax.set_xticks(ticks=[-0.2, 0.8, 1.8, 2.8])
ax.set_xticklabels(labels)
# beállítjuk az y tengely vonalat, ami alapján könnyen leolvashatóvá válik az ábra
ax.yaxis.grid(True)
ax.legend(loc='best')
ax.set_ylabel('Number of prizes')
fig.suptitle('Prizes by country')
fig.tight_layout(pad=2)
fig.savefig('mpl_barchart_multi.png', dpi=200)
```

```python
# pontok száma a diagramon
num_points = 100
# meredekség
gradient = 0.5

x = np.array(range(num_points))
# a randn 100 darab normális eloszlású véletlenszerű számokat ad, amely 0 és 10 között lesz, amihez még
# hozzáadunk egy x-től függő értéket is szórással
y = np.random.randn(num_points) * 10 + x * gradient

fig, ax = plt.subplots(figsize=(8, 4))

# meghatároz 100 darab random színértéket 0 és 1 között az alapértelmezett colormaphez
colors = np.random.rand(num_points)
# random méreteket határozunk meg a köreinknek 2 és 22 között, majd négyzetére emeljük, hogy variáljuk a méreteket
size = (2 + np.random.rand(num_points) * 20) ** 2

# létrehozzuk a szórtgrafikont
ax.scatter(x, y, s=size, c=colors, alpha=0.5)

fig.suptitle('A Simple Scatterplot with Color and Size specified')
```

```python
pal = dict(Female='red', Male='blue')

# a függvény betölt egy minta adatsort a Seaborn online repojából névszerint
tips = sns.load_dataset('tips')

# beadjuk az adatsort és az érdekelt oszlopot, amely alapján ábracsoportokat hoz létre
# a hue argumentum által meghatározunk egy jelölést a nem dimenziónak, a kws argumentum segítségével pedig,
# beállítjuk a gyémánt és négyzet jelöléseket, végül a palette-el a színt is
g = sns.FacetGrid(tips, col='smoker', hue='sex', hue_kws={'marker': ['D', 's']},
                 palette=pal, height=4, aspect=1)

# majd létrehozunk az ábracsoportok szerint különböző szórtdiagrammokat a teljes számla és
# borravaló arányából
g.map(plt.scatter, 'total_bill', 'tip', alpha=0.4)
g.add_legend()
```

```python
pal = dict(Female='red', Male='blue')

tips = sns.load_dataset('tips')

# beadjuk a 'time' sort is, hogy elkülönítsük a borravalókat idő szerint
g = sns.FacetGrid(tips, col='smoker', row='time', hue='sex', hue_kws={'marker': ['D', 's']},
                 palette=pal, height=4, aspect=1)

# a regplot regressziós görbét is elhelyez, nem szerinti konfidenciaintervallumra alapozva
g.map(sns.regplot, 'total_bill', 'tip')
g.add_legend()
```
